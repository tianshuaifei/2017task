```python
"""sentence char Length statistics"""
%matplotlib inline
import matplotlib.pyplot as plt
import json
import codecs
import jieba as jb
import os
import glob

#fig = plt.figure(figsize=(20,10))
fig = plt.figure()
# plt.title('Word length statistics', size=20)
# ax1 = fig.add_subplot(121)
# ax2 = fig.add_subplot(122)

# ax1.set_xlabel('word length', size=15)
# ax1.set_ylabel('num sentences', size=15)
# ax2.set_xlabel('word length', size=15)
# ax2.set_ylabel('num sentences', size=15)

char_files = glob.glob('./char/*.char')
word_files = glob.glob('./word/*.word')
tagSep = '\t'

count=0
stats={}
class_num=0
total_num=0
class_ = []

for name in char_files:
    with open(name, 'r') as fd:
        for line in fd:
            words = line.split()
            tags = words[0]
            newsItem = words[1:]
            if len(newsItem) not in stats:
                stats[len(newsItem)]=0
            stats[len(newsItem)]+=1
            total_num+=1
            for tag in tags.split(tagSep):
                if tag not in class_:
                    class_.append(tag)
                    class_num+=1
                
print "there are {} classes".format(class_num)
print "there are {} titles".format(total_num)
print class_

plt.plot([i for i in stats], [stats[i] for i in stats], label='Charactor level')
plt.legend()

count=0
stats={}
class_num=0
total_num=0
class_ = []

for name in word_files:
    with open(name, 'r') as fd:
        for line in fd:
            words = line.split()
            tags = words[0]
            newsItem = words[1:]
            if len(newsItem) not in stats:
                stats[len(newsItem)]=0
            stats[len(newsItem)]+=1
            total_num+=1
            for tag in tags.split(tagSep):
                if tag not in class_:
                    class_.append(tag)
                    class_num+=1
                
print "there are {} classes".format(class_num)
print "there are {} titles".format(total_num)
print class_

plt.plot([i for i in stats], [stats[i] for i in stats], label='Word level')
plt.legend()

plt.xlabel('Word length', size=10)
plt.ylabel('Num sentences', size=10)
plt.title('Word length statistics')

```
